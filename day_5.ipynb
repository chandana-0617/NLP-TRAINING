```python
""" Write a Python program to calculate the frequency of each word in a given text. Print the words and their corresponding counts"""


def calculate_word_frequency(text):
    words = text.lower().split()
    cleaned_words = [word.strip(",.?!;:'\"") for word in words]
    word_freq = {}
    for word in cleaned_words:
        word_freq[word] = word_freq.get(word, 0) + 1

    return word_freq

text = input("Enter a text: ")


word_frequency = calculate_word_frequency(text)

print("\nWord frequencies:")
for word, count in sorted(word_frequency.items()):
    print(f"{word}: {count}")
```
